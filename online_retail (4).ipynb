# Analyzing the Data:

#### Sales Trend:

```python
Sales_trend= retail_df.groupby(['InvoiceDate'], as_index = False)['InvoiceDate', 'Quantity'].sum()
new_df = Sales_trend.Quantity > 0  ## removing cancelled/returned orders
final_sales = Sales_trend[new_df] 
plt.plot(final_sales['InvoiceDate'], final_sales['Quantity'])
plt.figure(figsize=(100,5))
```

```python
retail_df['Year_Month'] = retail_df.Year.astype(str) + '_' + retail_df.Month.astype(str)
sales = retail_df.groupby(['Year_Month'], as_index = False)['Year_Month', 'Quantity'].sum()
sales.sort_values(['Year_Month'])

```

#### Weekday Sales Trend

```python
retail_df['Revenue'] = retail_df['Quantity'] * retail_df['UnitPrice']

retail_df['New_Date'] = 'Other'
retail_df.loc[retail_df['Day']== 0, 'New_Date'] = 'Monday'
retail_df.loc[retail_df['Day']== 1, 'New_Date'] = 'Tuesday'
retail_df.loc[retail_df['Day']== 2, 'New_Date'] = 'Wednesday'
retail_df.loc[retail_df['Day']== 3, 'New_Date'] = 'Thursday'
retail_df.loc[retail_df['Day']== 4, 'New_Date'] = 'Friday'
retail_df.loc[retail_df['Day']== 5, 'New_Date'] = 'Saturday'
retail_df.loc[retail_df['Day']== 6, 'New_Date'] = 'Sunday'


```

```python
weekday_sales = retail_df.groupby(['New_Date'], as_index = False)['New_Date', 'Revenue', 'Quantity'].sum()
weekday_sales.sort_values('Revenue', ascending = False)
```

#### Country-wise Analysis

```python
country = retail_df.groupby(['Country'], as_index = False).sum()
plt.bar(country['Country'], country['Quantity'])
plt.title('Country wise total quantity sold')
plt.xlabel('Countries')
plt.ylabel('Total Quantity')
plt.xticks(rotation=90)
plt.figure(figsize=(25,5))
```

```python
country['Proportion_totalqty'] = (country.Quantity/country.Quantity.sum())*100
```

```python
country.sort_values('Quantity', ascending = False).head(10)
```

#### The largest market share in terms of sales is United Kingdom which consist of around 82% of the total sales of the store.
