#### Checking for NaN Values in each attribute in data set: 

```python
print(retail_df.isna().sum())
```

```python
Nan_values1 = retail_df[retail_df['CustomerID'].astype(str).str.contains('n')]
Nan_values1
```

```python
p1 = Nan_values1.Quantity.sum()/retail_df.Quantity.sum()
print('Proportion of Quantity sold which has no information about Customer is {} ({:.2f}%)'.format(p1, p1*100))
p2 = Nan_values1.Quantity.count()/retail_df.Quantity.count()
print('Percentage of rows with no information about Customer is {} ({:.2f}%)'.format(p2, p2*100))
```

#### There are two ways to treat rows with NaN values. Either to keep the data or drop those rows. In our study, it is not beneficial to drop those rows because:
#### For around 1/4th of the rows, we have NaN value in customer ID. Dropping these rows might lead to data biasness. 

#### We will continue our analysis taking all these rows. 




#### Information about Quantitative Data: 

```python
retail_df.describe()
```

#### Important thing to notice is that there are some negative values in Quantity and UnitPrice. 
#### We can check if these quantities along with their UnitPrice are the ones that were returned or cancelled.

```python
cancelled_orders = retail_df[retail_df['InvoiceNo'].astype(str).str.contains('C')]
cancelled_orders
```

```python
cancelled_orders[cancelled_orders.Quantity > 0].Quantity.count()
```

```python
proportion = len(cancelled_orders)/len(retail_df.InvoiceNo.unique())
print('Percentage of cancelled orders is {} ({:.2f}%)'.format(proportion, proportion*100))
```

#### Hence, the negative values in Quantity column is basically the cancelled orders. However, we have around 36% of the total orders as cancelled orders which are huge. Dropping them will not be helpful. Rather, we can check in later stages, what all products have higher return rate to prevent future cancellation.

## Task 4: Analyzing the Data:

#### Sales Trend:

```python

plt.plot(retail_df.InvoiceDate, retail_df.Quantity)

```

#### Country-wise Analysis

```python
country = retail_df.groupby(['Country'], as_index = False).sum()
plt.bar(country['Country'], country['Quantity'])
plt.title('Country wise total quantity sold')
plt.xlabel('Countries')
plt.ylabel('Total Quantity')
plt.xticks(rotation=90)
plt.figure(figsize=(25,5))
```

```python
country['Proportion_totalqty'] = (country.Quantity/country.Quantity.sum())*100
```

```python
country.sort_values('Quantity', ascending = False).head(10)
```

#### The largest market share in terms of sales is United Kingdom which consist of around 82% of the total sales of the store.
